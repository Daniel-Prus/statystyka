```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

plt.rcParams['figure.figsize'] = (12,7)
```

```python
df = pd.read_csv('height_weight.csv')
df.head()
```

```python
m = df['sex'] == 1
```

```python
plt.scatter(df.loc[m,'height'], df.loc[m,'weight'], s=5, label='male')
plt.scatter(df.loc[~m,'height'], df.loc[~m,'weight'], s=5, label='female')
plt.xlabel('height')
plt.ylabel('weight')
plt.legend(loc=2);
```
