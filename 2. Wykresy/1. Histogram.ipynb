```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

#### 1. Data:

```python
d1 = np.loadtxt('example_1.txt')
d2 = np.loadtxt('example_2.txt')
df = pd.DataFrame(data={'d1': d1, 'd2':d2})
```

#### 2. Histogram - matplotlib + sns

```python
#count

bins = np.linspace(min(d1.min(), d2.min()), max(d1.max(), d2.max()), num=50)
plt.figure(figsize=(10,6))
plt.hist(d1, bins=bins, label='example_1')
plt.hist(d2, bins=bins, label='example_2', alpha=0.5)
plt.ylabel('Count')
plt.legend()
plt.show();
```

```python
# probability
# histtype - daje tylko obrys
# density to probability prawdopodobie≈Ñstwo)

bins = np.linspace(min(d1.min(), d2.min()), max(d1.max(), d2.max()), num=50)
plt.figure(figsize=(10,6))
plt.hist(d1, histtype='step', bins=bins, label='d1', density=True)
plt.hist(d2, histtype='step', bins=bins, label='d2', density=True, ls=":")
plt.ylabel('Probability')
plt.legend()
plt.show();
```

```python
# stacked 

# histtype="barstacked" (w stacked) powoduje co≈õ tam sprawd≈∫

bins = np.linspace(min(d1.min(), d2.min()), max(d1.max(), d2.max()), num=50)
plt.figure(figsize=(10,6))
plt.hist([d1, d2], bins=bins, label="Stacked", density=True, alpha=0.5)
plt.hist(d1, bins=bins, label="D1", density=True, histtype="step", lw=1)
plt.hist(d2, bins=bins, label="D2", density=True, histtype="step", ls=":")
plt.legend()
plt.ylabel("Probability");
```
