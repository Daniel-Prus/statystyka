```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (12,7)
sns.set()
```

```python
df = pd.read_csv('height_weight.csv')
df.head()
```

```python
sns.pairplot(df, corner=True);
```

```python
g = sns.pairplot(df, diag_kind="kde")
g.map_lower(sns.kdeplot, levels=4, color=".2");
            
```

```python
sns.pairplot(df, kind='kde', hue='sex', aspect=1.5, height=4.5, palette=['tab:blue','tab:red']);
```

```python
sns.pairplot(df, hue='sex', kind='reg', plot_kws={'ci':None, 'color':'salmon', 'scatter_kws':{'color':'gray'}})
```

```python
sns.pairplot(df, diag_kws={'color':'grey', 'alpha':1})
```
