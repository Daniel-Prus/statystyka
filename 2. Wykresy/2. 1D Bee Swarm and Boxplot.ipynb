```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (12,8)
sns.set()
```

```python
d1 = np.loadtxt('example_1.txt')
d2 = np.loadtxt('example_2.txt')
```

```python
bins= np.linspace(min(d1.min(), d2.min()), max(d1.max(), d2.max()), num=50)

plt.figure(figsize=(10,6))
plt.hist(d1, label='d2', bins=bins, density=True )
plt.hist(d2, label='d2', bins=bins, alpha=0.5, density=True)
plt.ylabel('Propability')
plt.legend()
plt.show()
```

### 1. Swarmplot()

```python
dataset = pd.DataFrame(data={
    'data': np.concatenate((d1,d2)),
    'type': np.concatenate((np.zeros(len(d1)), np.ones(len(d2)) ))
})

dataset['type']=dataset['type'].astype('int')
dataset.info()
```

```python
# size - musi być mniejsze (default-5), bo inaczej się wykrzaczy

sns.swarmplot(x='type',y='data', data=dataset, size=5);
```

### 2. Boxplot

- parametr whis - powoduje zmniejszenie ilości outliers (kontrola outliers)

```python
sns.swarmplot(x='type',y='data', data=dataset, size=5, color='grey', alpha=0.5)
sns.boxplot(x="type", y="data", data=dataset, whis=3.0);
```

### 3. Violinplot

- violinplot można także podzielić np. na male i female, trzeba sprawdzić dokumentację

```python
sns.violinplot(x="type", y="data", data=dataset)
sns.swarmplot(x='type',y='data', data=dataset, size=4, color='k', alpha=0.5);
```

```python
sns.violinplot(x="type", y="data", data=dataset, inner='quartile')
```

```python
sns.violinplot(x="type", y="data", data=dataset, inner='quartile', bw=0.1)
```
